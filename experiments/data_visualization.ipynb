```python
import pandas as pd

# Load dataset
disease_df_train = pd.read_csv('../data/processed/Training.csv')

# Display dataset
disease_df_train.head()
```

```python
import matplotlib.pyplot as plt

# Summing the symptoms to get their frequency
symptom_counts = disease_df_train.iloc[:, :-1].sum()

# Plotting the frequency distribution
symptom_counts.plot(kind='bar', figsize=(20, 10))
plt.title('Frequency Distribution of Symptoms')
plt.xlabel('Symptoms')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend(title='Count', loc='upper right')
plt.show()
```

```python
# Get prognosis values
unique_prognosis = disease_df_train['prognosis'].unique()

# Iterate over prognosis values and plot frequency of distribution per symptom
for prognosis in unique_prognosis:
    # Filter DataFrame for current prognosis
    filtered_data = disease_df_train[disease_df_train['prognosis'] == prognosis]
    
    # Calculate symptom counts
    symptom_counts = filtered_data.iloc[:, :-1].sum()

    # Plot
    symptom_counts.plot(kind='bar', figsize=(20, 10))
    plt.title(f'Frequency Distribution of Symptoms for {prognosis}')
    plt.xlabel('Symptoms')
    plt.ylabel('Frequency')
    plt.xticks(rotation=90)
    plt.legend(title='Count', loc='upper right')
    plt.show()
```
