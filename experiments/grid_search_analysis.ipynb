```python
import os
import pandas as pd

# Specify the directory containing the CSV files
directory_path = '../reports/'  # Update with your directory path

# Get a list of all CSV files in the directory
csv_files = [f for f in os.listdir(directory_path) if f.endswith('.csv')]

# Initialize an empty list to store the DataFrames
all_results = []

# Loop through each CSV file, read it, and append to the list
for file in csv_files:
    file_path = os.path.join(directory_path, file)
    df = pd.read_csv(file_path)
    df['source_file'] = file  # Add the source file name as a new column for reference
    all_results.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_results = pd.concat(all_results, ignore_index=True)

print(combined_results.columns)

# Show the first few rows of the combined results
# print(combined_results.head())

# Optionally, save the combined results to a new CSV file
# combined_results.to_csv('./reports/combined_grid_search_results.csv', index=False)
```

```python
# Example: Find the best score for each source file (model)
best_results = combined_results.loc[combined_results.groupby('source_file')['mean_test_accuracy'].idxmax()]
print(best_results)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: Boxplot for performance across batch sizes
plt.figure(figsize=(10, 6))
sns.boxplot(x='param_classifier__batch_size', y='mean_test_score', data=combined_results)
plt.title("Performance vs. Batch Size")
plt.xlabel("Batch Size")
plt.ylabel("Mean Test Score")
plt.show()
```
