```python
def list_symptoms_columns(disease_df_train, target_column='prognosis'):
    """
    Lists all columns in the DataFrame except the target column.
    
    Args:
    dataframe (pd.DataFrame): The input DataFrame.
    target_column (str): The name of the target column to exclude (default is 'prognosis').
    
    Returns:
    list: A list of symptom column names.
    """
    # Get all columns except the target column
    symptoms = [col for col in disease_df_train.columns if col != target_column]
    
    # Print the result
    print("List of symptoms:")
    print(symptoms)
    
    return symptoms

# Call the function and store the result
symptom_columns = list_symptoms_columns(disease_df_train)
```

```python
# Identify the target column (assumed to be the last column)
disease_column = disease_df_train.columns[-1]


# Group symptoms by disease and calculate their mean occurrence
symptoms_by_disease = disease_df_train.groupby(disease_column).mean()

# Create a dictionary to store minimum unique symptoms for each disease
unique_symptoms_per_disease = {}

# Iterate through each disease to identify unique symptoms
for disease in symptoms_by_disease.index:
    # For the current disease, find symptoms with high prevalence (e.g., > 0.5)
    significant_symptoms = symptoms_by_disease.loc[disease][
        symptoms_by_disease.loc[disease] > 0.5
    ].index.tolist()
    
    # Store in dictionary
    unique_symptoms_per_disease[disease] = significant_symptoms

# Display the dictionary
print("Unique Symptoms Per Disease:")
print(unique_symptoms_per_disease)
```

## Regression Analyses 

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# Read and prepare the data
data = pd.read_csv('../data/processed/Training.csv')

data.head()

# Separate features and target
X = data.iloc[:, :-1]  # All columns except the last one
y = data.iloc[:, -1]   # Last column (prognosis)

# Encode the target variable
le = LabelEncoder()
y = le.fit_transform(y)

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, 
                                                    test_size=0.2, 
                                                    random_state=42, 
                                                    stratify=y)

# Create model with regularization
model = LogisticRegression(
    C=0.1,                # Stronger regularization
    max_iter=1000,        # More iterations for convergence
    class_weight='balanced'  # Handle class imbalance
)
```

```python
# Perform cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X_train, y_train, cv=cv)

print("Cross-validation scores:", cv_scores)
print(f"Average CV Score: {cv_scores.mean():.2f} (+/- {cv_scores.std() * 2:.2f})")

# Train the final model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Print detailed metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
# Get disease names
disease_names = le.classes_

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Create figure and plot
plt.figure(figsize=(15, 12))

# Create heatmap
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=disease_names,
            yticklabels=disease_names)

# Rotate labels for better readability
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Add labels and title
plt.title('Confusion Matrix of Disease Predictions', pad=20)
plt.ylabel('True Disease')
plt.xlabel('Predicted Disease')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show plot
plt.show()

print("\nDisease Classes:")
for i, disease in enumerate(disease_names):
    print(f"{i}: {disease}")
```

```python
# Calculate feature importance with confidence intervals
n_bootstraps = 10
feature_importance_bootstrapped = np.zeros((n_bootstraps, X.shape[1]))

for i in range(n_bootstraps):
    # Bootstrap sample indices
    indices = np.random.randint(0, len(X_train), len(X_train))
    # Fit model on bootstrap sample
    model_boot = LogisticRegression(C=0.1, max_iter=1000)
    model_boot.fit(X_train[indices], y_train[indices])
    # Store feature importance
    feature_importance_bootstrapped[i] = abs(model_boot.coef_[0])

# Calculate mean and confidence intervals
feature_importance_mean = np.mean(feature_importance_bootstrapped, axis=0)
feature_importance_ci = np.percentile(feature_importance_bootstrapped, [2.5, 97.5], axis=0)

# Create feature importance DataFrame
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importance_mean,
    'CI_lower': feature_importance_ci[0],
    'CI_upper': feature_importance_ci[1]
})

print("\nTop 10 Most Important Features with 95% Confidence Intervals:")
top_features = feature_importance.sort_values('Importance', ascending=False).head(10)
print(top_features[['Feature', 'Importance', 'CI_lower', 'CI_upper']])
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# Find best K value
k_range = range(1, 31)
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    scores.append(knn.score(X_test, y_test))

# Plot accuracy vs K value
plt.figure(figsize=(10, 6))
plt.plot(k_range, scores, 'bx-')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.title('Accuracy vs K Value')
plt.grid(True)
plt.show()

# Get best K value
best_k = k_range[np.argmax(scores)]
print(f"\nBest K value: {best_k}")

# Train model with best K
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train, y_train)

# Make predictions
y_pred = best_knn.predict(X_test)

# Print classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, 
                          target_names=le.classes_))

# Create confusion matrix
plt.figure(figsize=(15, 12))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_,
            yticklabels=le.classes_)

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.title('Confusion Matrix (KNN Classification)', pad=20)
plt.ylabel('True Disease')
plt.xlabel('Predicted Disease')
plt.tight_layout()
plt.show()

# Compare with nearest neighbors for a sample case
sample_idx = 0  # Taking first test case as example
sample_prediction = best_knn.predict(X_test[sample_idx:sample_idx+1])
neighbors = best_knn.kneighbors(X_test[sample_idx:sample_idx+1], 
                               return_distance=True)

print("\nSample Prediction Analysis:")
print(f"Actual disease: {le.classes_[y_test[sample_idx]]}")
print(f"Predicted disease: {le.classes_[sample_prediction[0]]}")
print("\nNearest Neighbors:")
for i, (distance, idx) in enumerate(zip(neighbors[0][0], neighbors[1][0])):
    print(f"Neighbor {i+1}: {le.classes_[y_train[idx]]} (Distance: {distance:.3f})")
```

## Neural Network 

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Read the data with new path
data = pd.read_csv('../data/processed/Training.csv')

# Separate features and target
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

# Encode the target variable
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_categorical = to_categorical(y_encoded)  # Convert to one-hot encoding

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_categorical,
                                                    test_size=0.2,
                                                    random_state=42)

# Create the model with specified architecture
model = Sequential()

# Add layers
model.add(Dense(32, input_dim=X_train.shape[1], activation='relu'))  # input layer + first hidden layer
model.add(Dense(16, activation='relu'))  # second hidden layer
model.add(Dense(41, activation='softmax'))  # output layer (41 classes = disease)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Summary of the model
model.summary()

# Train the model
history = model.fit(X_train, y_train,
                   epochs=50,
                   batch_size=32,
                   validation_split=0.2,
                   verbose=1)

# Plot training history
plt.figure(figsize=(12, 4))

# Loss plot
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Accuracy plot
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"\nTest Accuracy: {test_accuracy:.4f}")
```

```python
# Get feature names
feature_names = X.columns

# Function to calculate feature importance using permutation importance
def calculate_feature_importance(model, X, y, feature_names):
    baseline_score = model.evaluate(X, y, verbose=0)[1]
    importance_scores = []
    
    for column in range(X.shape[1]):
        # Store original values
        hold = X[:, column].copy()
        
        # Permute the feature
        X[:, column] = np.random.permutation(X[:, column])
        
        # Calculate new score
        permuted_score = model.evaluate(X, y, verbose=0)[1]
        
        # Calculate importance
        importance = baseline_score - permuted_score
        
        # Restore original values
        X[:, column] = hold
        
        importance_scores.append(importance)
    
    # Create DataFrame with feature names and importance scores
    feature_importance = pd.DataFrame({
        'Feature': feature_names,
        'Importance': importance_scores
    })
    
    return feature_importance.sort_values('Importance', ascending=False)

# Calculate feature importance
importance_df = calculate_feature_importance(model, X_scaled, y_categorical, feature_names)

# Display top 10 features
print("\nTop 10 Most Important Features:")
print(importance_df.head(10))

# Plot top 10 features
plt.figure(figsize=(12, 6))
plt.bar(importance_df['Feature'][:10], importance_df['Importance'][:10])
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Most Important Features')
plt.xlabel('Features')
plt.ylabel('Importance Score')
plt.tight_layout()
plt.show()

# Save feature importance results
importance_df.to_csv('feature_importance.csv', index=False)
```

Comparing the feature importances from logistic regression and neural networks 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create DataFrames for both models' results
nn_features = pd.DataFrame({
    'Feature': ['nodal_skin_eruptions', 'sunken_eyes', 'dehydration', 'dischromic_patches', 
                'altered_sensorium', 'coma', 'passage_of_gases', 'dark_urine', 
                'mild_fever', 'internal_itching'],
    'Importance': [0.002846, 0.002642, 0.002642, 0.002439, 0.002236, 
                   0.001423, 0.001423, 0.001423, 0.001220, 0.001220],
    'Model': 'Neural Network'
})

lr_features = pd.DataFrame({
    'Feature': ['unsteadiness', 'spinning_movements', 'loss_of_balance', 'nausea', 
                'headache', 'vomiting', 'chest_pain', 'dizziness', 'yellowish_skin', 
                'loss_of_appetite'],
    'Importance': [0.568765, 0.448178, 0.386381, 0.206106, 0.194695, 
                   0.128240, 0.072602, 0.070778, 0.065918, 0.064852],
    'Model': 'Logistic Regression'
})

# Normalize importance scores within each model
nn_features['Normalized_Importance'] = nn_features['Importance'] / nn_features['Importance'].max()
lr_features['Normalized_Importance'] = lr_features['Importance'] / lr_features['Importance'].max()

# Combine all features
all_features = pd.concat([
    nn_features[['Feature', 'Normalized_Importance', 'Model']],
    lr_features[['Feature', 'Normalized_Importance', 'Model']]
])

# Create a feature categorization
symptom_categories = {
    'Neurological': ['unsteadiness', 'spinning_movements', 'loss_of_balance', 'dizziness', 
                     'altered_sensorium', 'coma'],
    'Digestive': ['nausea', 'vomiting', 'passage_of_gases', 'dark_urine'],
    'Physical': ['nodal_skin_eruptions', 'dischromic_patches', 'sunken_eyes'],
    'Systemic': ['mild_fever', 'dehydration', 'headache', 'loss_of_appetite', 'yellowish_skin'],
    'Other': ['internal_itching', 'chest_pain']
}

# Add category to features
def get_category(feature):
    for category, symptoms in symptom_categories.items():
        if feature in symptoms:
            return category
    return 'Other'

all_features['Category'] = all_features['Feature'].apply(get_category)

# Create visualization
plt.figure(figsize=(15, 8))
sns.scatterplot(data=all_features, x='Feature', y='Normalized_Importance', 
                hue='Model', style='Category', s=100)
plt.xticks(rotation=45, ha='right')
plt.title('Combined Feature Importance Analysis')
plt.tight_layout()
plt.show()

# Print summary of combined insights
print("\nCombined Feature Analysis by Category:")
print("=====================================")
for category in symptom_categories.keys():
    category_features = all_features[all_features['Category'] == category]
    print(f"\n{category} Symptoms:")
    for model in ['Neural Network', 'Logistic Regression']:
        model_features = category_features[category_features['Model'] == model]
        if not model_features.empty:
            print(f"{model}:", ', '.join(model_features['Feature'].tolist()))
```
