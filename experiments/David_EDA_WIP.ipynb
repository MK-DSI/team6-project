### Analysis of symptoms

```python
# Creating a symptoms df
symptom_columns = disease_df_train.columns[:-1]  # Removing the prognosis column 
symptoms_df = disease_df_train[symptom_columns]

symptoms_df.head()
```

```python
def list_symptoms_columns(disease_df_train, target_column='prognosis'):
    """
    Lists all columns in the DataFrame except the target column.
    
    Args:
    dataframe (pd.DataFrame): The input DataFrame.
    target_column (str): The name of the target column to exclude (default is 'prognosis').
    
    Returns:
    list: A list of symptom column names.
    """
    # Get all columns except the target column
    symptoms = [col for col in disease_df_train.columns if col != target_column]
    
    # Print the result
    print("List of symptoms:")
    print(symptoms)
    
    return symptoms

# Call the function and store the result
symptom_columns = list_symptoms_columns(disease_df_train)
```

```python
num_symptoms_per_patient = disease_df_train[symptom_columns].sum(axis=1)
print(num_symptoms_per_patient)
# Use Pandas methods for mean, max, and min
print(f"Mean symptoms per patient: {num_symptoms_per_patient.mean()}")
print(f"Max symptoms per patient: {num_symptoms_per_patient.max()}")
print(f"Min symptoms per patient: {num_symptoms_per_patient.min()}")

```

```python
# Calculate the frequency of each symptom
symptom_frequencies = symptoms_df.sum(axis=0)

print(symptom_frequencies)
```

```python
# Create a DataFrame for better visualization
symptom_frequencies_df = pd.DataFrame({
    "Symptom": symptom_frequencies.index,
    "Frequency": symptom_frequencies.values
}).sort_values(by="Frequency", ascending=False)

# Extract the top 10 most and least frequent symptoms
top_10_symptoms = symptom_frequencies_df.head(10)
least_10_symptoms = symptom_frequencies_df.tail(10)

# Display the results
print("Top 10 Most Frequent Symptoms:")
print(top_10_symptoms)
print("\nTop 10 Least Frequent Symptoms:")
print(least_10_symptoms)
```

#### Examining the correlation of symptoms 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Numeric symptoms represented in the symptoms df 
# symptoms_df.head()

# Calculate the correlation matrix for symptoms
correlation_matrix = symptoms_df.corr()

# Visualize the correlation matrix with a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False, cbar=True)
plt.title("Correlation Matrix of Symptoms")
plt.show()
```

```python
# Extract pairs of symptoms with high correlations (e.g., > 0.8)
highly_correlated_pairs = correlation_matrix.where(
    (correlation_matrix > 0.8) & (correlation_matrix < 1.0)
).stack().reset_index()

print(highly_correlated_pairs.columns)
```

```python
# Rename columns for clarity
highly_correlated_pairs.columns = ["Symptom 1", "Symptom 2", "Correlation"]
```

```python
# Sort the pairs by correlation in descending order
highly_correlated_pairs = highly_correlated_pairs.sort_values(by="Correlation", ascending=False)

# Display the first 10 highly correlated symptom pairs
print("Top 10 Highly Correlated Symptom Pairs:")
print(highly_correlated_pairs.head(10))
print(len(highly_correlated_pairs))
```
